## Step 5: Mark HMT as visited and update the table
**Remember:** The cost of VRT was calculated before and was 218 when coming from RBY. Now, coming from HMT, we get 253. We compare the costs and find that 253 is higher than 218. Although VRT is the next node, we will continue with the cost of 218.

## Step 3: Calculate the cost of JHB from VRT
- Cost of JHB = Cost of VRT + Distance from VRT to JHB  
  = 218 + 106  
  = 324

## Step 5: Mark VRT as visited and update the table accordingly

| Node | Cost | Previous Node |
|------|------|---------------|
| DBN  | 0    | -             |
| PMB  | 89   | DBN           |
| RBY  | 112  | DBN           |
| HMT  | 212  | RBY           |
| VRT  | 218  | RBY           |
| JHB  | 324  | VRT           |

Finally, determine the shortest path referring to the table:
- JHB - previous node was VRT
- VRT - previous node was RBY
- RBY - previous node was DBN

**Therefore:** The shortest path from DBN to JHB is DBN -> RBY -> VRT -> JHB with a total cost of R324.


```python
def find_shortest_path(graph, start, end):
    next_node = [(0, start, [])]
    visited = set()
    min_dist = {start: 0}

    while next_node:
        # Find the node with the smallest distance (cost)
        next_node.sort(key=lambda x: x[0])
        cost, node, path = next_node.pop(0)
        
        if node in visited:
            continue

        visited.add(node)
        path = path + [node]

        if node == end:
            return (cost, path)

        for neighbor, weight in graph[node].items():
            if neighbor in visited:
                continue
            old_cost = min_dist.get(neighbor, float('inf'))
            new_cost = cost + weight
            if new_cost < old_cost:
                min_dist[neighbor] = new_cost
                next_node.append((new_cost, neighbor, path))

    return (float('inf'), [])

# Define the nodes with their distances 
graph = {
    'DBN': {'PMB': 89, 'RBY': 112},
    'PMB': {'HMT': 209, 'RBY': 70},
    'RBY': {'HMT': 100, 'VRT': 106},
    'HMT': {'VRT': 41, 'JHB': 210},
    'VRT': {'JHB': 106},
    'JHB': {}
}

start = 'DBN'
end = 'JHB'

# Call the function find the shortest path
cost, path = find_shortest_path(graph, start, end)

# Output the results
print(f"The shortest path from {start} to {end} is {path} with a total cost of R{cost}.")
```
